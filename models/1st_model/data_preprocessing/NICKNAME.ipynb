```python
ns=[]
for i in range(5):
  df = pd.read_csv(f'/content/data/agg_match_stats_{i}.csv')
  names = df['player_name'].unique()
  df['player_name'] = df['player_name'].astype('category')
  name_counts = df['player_name'].value_counts()
  ns.append(name_counts)
```

```python
ns
```

```python
combined_series = pd.concat(ns)

# 2. ÏïÑÏù¥Îîî(index)Î•º Í∏∞Ï§ÄÏúºÎ°ú Í∑∏Î£πÌôî(groupby)ÌïòÍ≥† Ìï©Í≥Ñ(sum)Î•º Íµ¨Ìï©ÎãàÎã§.
total_counts = combined_series.groupby(level=0).sum()

# 3. (ÏÑ†ÌÉù) Ìï©Í≥ÑÍ∞Ä ÎÜíÏùÄ ÏàúÏúºÎ°ú Ï†ïÎ†¨ÌïòÏó¨ Í≤∞Í≥ºÎ•º ÌôïÏù∏Ìï©ÎãàÎã§.
sorted_counts = total_counts.sort_values(ascending=False)

print(sorted_counts)
```

```python
# 'player_counts.csv' ÎùºÎäî Ïù¥Î¶ÑÏùò ÌååÏùºÎ°ú Ï†ÄÏû•Ìï©ÎãàÎã§.
# header=True ÏòµÏÖòÏúºÎ°ú Ïó¥ Ï†úÎ™©ÏùÑ Ï∂îÍ∞ÄÌï¥Ï£ºÎäî Í≤ÉÏù¥ Ï¢ãÏäµÎãàÎã§.
sorted_counts.to_csv('player_counts.csv', header=True)

print("'player_counts.csv' ÌååÏùº Ï†ÄÏû•Ïù¥ ÏôÑÎ£åÎêòÏóàÏäµÎãàÎã§.")
```

```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm # tqdm ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞

# ======================================================================
# 1Îã®Í≥Ñ: ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∞è Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
# ======================================================================
try:
    df_players = pd.read_csv('/content/player_counts_40.csv')
    player_names = df_players['player_name'].unique()
    print(f"Î∂ÑÏÑù ÎåÄÏÉÅ ÌîåÎ†àÏù¥Ïñ¥: Ï¥ù {len(player_names)}Î™Ö")
except FileNotFoundError:
    print("Ïò§Î•ò: '/content/player_counts.csv' ÌååÏùºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§. ÌååÏùº Í≤ΩÎ°úÎ•º ÌôïÏù∏Ìï¥Ï£ºÏÑ∏Ïöî.")
    player_names = []

stat_cols = [
    'player_assists', 'player_dbno', 'player_dist_ride', 'player_dist_walk',
    'player_dmg', 'player_kills', 'player_survive_time', 'team_placement'
]

if len(player_names) > 0:
    total_stats = pd.DataFrame(0, index=player_names, columns=stat_cols, dtype=np.float64)
    total_games = pd.Series(0, index=player_names, dtype=int)

# ======================================================================
# 2Îã®Í≥Ñ: 5Í∞ú Ïä§ÌÉØ ÌååÏùº ÏàúÌöå Î∞è Îç∞Ïù¥ÌÑ∞ Ìï©ÏÇ∞ (tqdm Î∞è Ïù¥ÏÉÅÏπò Ï†úÍ±∞ Ï†ÅÏö©)
# ======================================================================
if len(player_names) > 0:
    for i in tqdm(range(5), desc="üìä Ïä§ÌÉØ ÌååÏùº Ï≤òÎ¶¨ Ï§ë"):
        file_path = f'/content/data/agg_match_stats_{i}.csv'
        try:
            df_stats_part = pd.read_csv(file_path)

            df_stats_part = df_stats_part[
                (df_stats_part['player_kills'] < 35) &
                (df_stats_part['player_dmg'] < 4000) &
                (df_stats_part['player_dbno'] < 35) &
                (df_stats_part['player_assists'] < 20) &
                (df_stats_part['player_dist_walk'] < 7000) &
                (df_stats_part['player_dist_ride'] < 35000) &
                (df_stats_part['player_survive_time'] < 35000) &
                (df_stats_part['team_placement'] != 0)&
                (df_stats_part['team_placement'] <= 100)
            ]

            filtered_df = df_stats_part[df_stats_part['player_name'].isin(player_names)]

            if filtered_df.empty:
                continue

            games_in_file = filtered_df.groupby('player_name').size()
            stats_in_file = filtered_df.groupby('player_name')[stat_cols].sum()

            total_games = total_games.add(games_in_file, fill_value=0)
            total_stats = total_stats.add(stats_in_file, fill_value=0)
        except FileNotFoundError:
            tqdm.write(f"Í≤ΩÍ≥†: '{file_path}' ÌååÏùºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏñ¥ Í±¥ÎÑàÎúÅÎãàÎã§.")

    print("\n‚úÖ Î™®Îì† ÌååÏùº Ï≤òÎ¶¨ Î∞è Ìï©ÏÇ∞ ÏôÑÎ£å!")

# ======================================================================
# 3Îã®Í≥Ñ: ÏµúÏ¢Ö ÌèâÍ∑† Í≥ÑÏÇ∞
# ======================================================================
if len(player_names) > 0 and not total_stats.empty:
    total_games_no_zero = total_games.replace(0, 1)
    average_stats = total_stats.div(total_games_no_zero, axis=0)
    average_stats[total_games == 0] = 0

    print("\n## üìä ÏµúÏ¢Ö ÌèâÍ∑† Ïä§ÌÉØ Í≤∞Í≥º (ÏÉÅÏúÑ 5Î™Ö) ##")
    print(average_stats.head())
else:
    print("\nÏ≤òÎ¶¨Ìï† Îç∞Ïù¥ÌÑ∞Í∞Ä ÏóÜÏñ¥ ÌèâÍ∑† Í≥ÑÏÇ∞ÏùÑ Í±¥ÎÑàÎúÅÎãàÎã§.")
```

```python
import pandas as pd
import glob

# 1. ÌååÏùºÎì§Ïù¥ Ï†ÄÏû•Îêú Í≤ΩÎ°úÎ•º ÏßÄÏ†ïÌï©ÎãàÎã§.
# ÏΩîÎû©Ïùò '/content/data/' Ìè¥ÎçîÏóê ÌååÏùºÏù¥ ÏûàÎã§Í≥† Í∞ÄÏ†ïÌï©ÎãàÎã§.
path = '/content/data/'

# 2. globÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Ìï¥Îãπ Í≤ΩÎ°úÏóêÏÑú 'agg_match_stats_*.csv' Ìå®ÌÑ¥ÏùÑ Í∞ÄÏßÑ Î™®Îì† ÌååÏùºÏùò Î™©Î°ùÏùÑ Í∞ÄÏ†∏ÏòµÎãàÎã§.
# '*'Îäî Ïñ¥Îñ§ Î¨∏Ïûê/Ïà´ÏûêÍ∞Ä ÏôÄÎèÑ Ï¢ãÎã§Îäî ÏùòÎØ∏Ïùò ÏôÄÏùºÎìúÏπ¥ÎìúÏûÖÎãàÎã§.
file_list = glob.glob(path + "agg_match_stats_*.csv")

# ÌååÏùºÏù¥ Ï†úÎåÄÎ°ú Ï∞æÏïÑÏ°åÎäîÏßÄ ÌôïÏù∏ (ÏÑ†ÌÉù ÏÇ¨Ìï≠)
print("Ï∞æÏïÑÎÇ∏ ÌååÏùº Î™©Î°ù:")
print(file_list)

# 3. Í∞Å ÌååÏùºÏùÑ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú ÏùΩÏñ¥ÏôÄ Î¶¨Ïä§Ìä∏Ïóê Îã¥ÏäµÎãàÎã§.
df_list = [pd.read_csv(file) for file in file_list]

# 4. pd.concatÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Î¶¨Ïä§Ìä∏Ïóê ÏûàÎäî Î™®Îì† Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ÌïòÎÇòÎ°ú Ìï©Ïπ©ÎãàÎã§.
# ignore_index=True ÏòµÏÖòÏùÄ Í∏∞Ï°¥ ÌååÏùºÎì§Ïùò Ïù∏Îç±Ïä§Î•º Î¨¥ÏãúÌïòÍ≥† ÏÉàÎ°ú Ïù∏Îç±Ïä§Î•º Î∂ÄÏó¨Ìï©ÎãàÎã§.
combined_df = pd.concat(df_list, ignore_index=True)

# 5. ÏµúÏ¢Ö Í≤∞Í≥ºÎ•º ÌôïÏù∏Ìï©ÎãàÎã§.
print("\n‚úÖ Î™®Îì† ÌååÏùºÏù¥ ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Ìï©Ï≥êÏ°åÏäµÎãàÎã§.")
print("Ìï©Ï≥êÏßÑ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò ÌÅ¨Í∏∞ (Ìñâ, Ïó¥):", combined_df.shape)
print("\nÌï©Ï≥êÏßÑ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÎØ∏Î¶¨Î≥¥Í∏∞:")
print(combined_df.head())
```
