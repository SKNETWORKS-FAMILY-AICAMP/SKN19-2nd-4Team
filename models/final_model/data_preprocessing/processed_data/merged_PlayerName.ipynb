```python
import pandas as pd 

df = pd.read_csv('./data/average_stats_erangel.csv')
df0 = pd.read_csv('./data/kill_match_stats_final_0.csv')
df1 = pd.read_csv('./data/kill_match_stats_final_1.csv')
df2 = pd.read_csv('./data/kill_match_stats_final_2.csv')
df3 = pd.read_csv('./data/kill_match_stats_final_3.csv')
df4 = pd.read_csv('./data/kill_match_stats_final_4.csv')

df0_killer = df0.rename(columns={'killer_name' : 'player_name'}) 
merged_by_killer = pd.merge(df,df0_killer, on='player_name', how='inner')
merged_by_killer.to_csv('merged_by_killer0.csv', index=False)

df0_victim = df0.rename(columns={'victim_name' : 'player_name'})
merged_by_victim = pd.merge(df, df0_victim, on = 'player_name', how='inner')
merged_by_victim.to_csv('merged_by_victim0.csv', index=False)

df1_killer = df1.rename(columns={'killer_name' : 'player_name'}) 
merged_by_killer = pd.merge(df,df1_killer, on='player_name', how='inner')
merged_by_killer.to_csv('merged_by_killer1.csv', index=False)

df1_victim = df1.rename(columns={'victim_name' : 'player_name'})
merged_by_victim = pd.merge(df, df1_victim, on = 'player_name', how='inner')
merged_by_victim.to_csv('merged_by_victim1.csv', index=False)

df2_killer = df2.rename(columns={'killer_name' : 'player_name'}) 
merged_by_killer = pd.merge(df,df2_killer, on='player_name', how='inner')
merged_by_killer.to_csv('merged_by_killer2.csv', index=False)

df3_victim = df3.rename(columns={'victim_name' : 'player_name'})
merged_by_victim = pd.merge(df, df3_victim, on = 'player_name', how='inner')
merged_by_victim.to_csv('merged_by_victim3.csv', index=False)

df3_killer = df3.rename(columns={'killer_name' : 'player_name'}) 
merged_by_killer = pd.merge(df,df3_killer, on='player_name', how='inner')
merged_by_killer.to_csv('merged_by_killer3.csv', index=False)

df4_victim = df4.rename(columns={'victim_name' : 'player_name'})
merged_by_victim = pd.merge(df, df4_victim, on = 'player_name', how='inner')
merged_by_victim.to_csv('merged_by_victim4.csv', index=False)

df4_killer = df4.rename(columns={'killer_name' : 'player_name'}) 
merged_by_killer = pd.merge(df,df4_killer, on='player_name', how='inner')
merged_by_killer.to_csv('merged_by_killer4.csv', index=False)

# df = pd.read_csv('./data/average_stats_erangel.csv')
# df0 = pd.read_csv('./data/kill_match_stats_final_4.csv')

# df0_killer = df0.rename(columns={'killer_name' : 'player_name'}) 
# merged_by_killer = pd.merge(df,df0_killer, on='player_name', how='inner')
# merged_by_killer.to_csv('merged_by_killer0.csv', index=False)

# df0_victim = df0.rename(columns={'victim_name' : 'player_name'})
# merged_by_victim = pd.merge(df, df0_victim, on = 'player_name', how='inner')
# merged_by_victim.to_csv('merged_by_victim0.csv', index=False)


file_list = ['merged_by_victim0.csv','merged_by_victim1.csv','merged_by_victim2.csv','merged_by_victim3.csv','merged_by_victim4.csv']

df_list = [pd.read_csv(fname) for fname in file_list]
merged = pd.concat(df_list, ignore_index=True)

# merged = pd.concat([df1,df2], ignore_index=True)

filtered = merged[merged['time'] <= 420]
filtered.to_csv('merged_victim_filtered_time_420.csv', index=False)
```
